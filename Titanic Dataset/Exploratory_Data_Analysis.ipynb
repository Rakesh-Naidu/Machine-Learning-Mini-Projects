## 5th Step - Outliers Treatment

```python
data.boxplot('Age')
```

```python
data.boxplot("Fare")
```

```python
data.boxplot("SibSp")
```

```python
data.boxplot("Parch")
```

### Age

```python
data['Age'].describe()
```

```python
IQR = data['Age'].quantile(0.75)-data['Age'].quantile(0.25)
print(IQR)
```

```python
Upper= data['Age'].quantile(0.75)+1.5*IQR
Lower=  data['Age'].quantile(0.25)-1.5*IQR
print(Upper)
print(Lower)
```

#### Lower Outliers

```python
Lower_Outliers=data[(data['Age']<=Lower)]
Lower_Outliers
```

```python
lis=Lower_Outliers.index
lis
```

```python
for i in lis:
    data.loc[i,'Age']=data['Age'].quantile(0.05)
```

```python
data.boxplot('Age')
```

#### Upper Outliers

```python
Upper_Outliers=data[(data['Age']>=Upper)]
Upper_Outliers
```

```python
lis=Upper_Outliers.index
lis
```

```python
for i in lis:
    data.loc[i,'Age']=data['Age'].quantile(0.95)
```

```python
data.boxplot('Age')
```
