# Machine Learning On Titanic Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data=pd.read_csv("final_Titanic.csv")
```

### Defining X and Y

```python
data
```

```python
X=data.iloc[:,1:].values
Y=data.iloc[:,0:1].values
X
```

```python
Y
```

```python
X.shape
```

```python
Y.shape
```

## Splitting The data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

# Building Models

## 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,Y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
Y_test
```

```python
y_pred
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(Y_test,y_pred)
confusion
```

```python
TN = confusion[0,0]
FP = confusion[0,1]
FN = confusion[1,0]
TP = confusion[1,1]
print("True Negative: ",TN)
print("False Positive: ",FP)
print("False Negative: ",FN)
print("True Positive: ",TP)
```

```python
cf=pd.DataFrame(confusion)
cf.columns=['Predicted Died','Predicted Survived']
cf= cf.rename(index = {0:'Actual Died',1:'Actual Survived'})
cf
```

### Confusion Matrix Evaluation Metrics

#### Classification Accuracy

```python
from sklearn import metrics
log = metrics.accuracy_score(Y_test,y_pred)
print(round(log,4))
```

#### Classification Error

```python
print(round(1-log,4))
```

### Sensitivity

#### Recall

```python
print("RECALL/TRUE POSITIVE RATE: ",metrics.recall_score(Y_test,y_pred))
```

#### Specificity

```python
print("SPECIFICITY/TRUE NEGATIVE RATE: ",(TN)/(TN+FP))
```

#### False Positive Rate

```python
print("FALSE POSITIVE RATE: ",(FN)/(FN+TP))
```

#### False Negative Rate

```python
print("FALSE NEGATIVE RATE: ",(FP)/(TN+FP))
```

#### Precision

```python
print("PRECISION: ",round(metrics.precision_score(Y_test,y_pred),2))
```

#### F1 Score

```python
print("Accuracy is: ",round(log,2)*100)
print("F1 Score is: ",round(metrics.f1_score(Y_test,y_pred),2)*100)
print("Precision is: ",round(metrics.precision_score(Y_test,y_pred),2)*100)
print("Recall is: ",round(metrics.recall_score(Y_test,y_pred),2)*100)
print("Roc Auc is: ",round(metrics.roc_auc_score(Y_test,y_pred),2)*100)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_pred))
```

## 3. Random Forest

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 40, n_jobs= 2, random_state= 0)
clf.fit(X_train, Y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
Y_test
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_pred)
cm
```

```python
rf=metrics.accuracy_score(Y_test, y_pred)
print("Accuracy:",round(metrics.accuracy_score(Y_test, y_pred)*100,2),"%")
```

```python
cf=pd.DataFrame(cm)
cf.columns=['Predicted Died','Predicted Survived']
cf= cf.rename(index = {0:'Actual Died',1:'Actual Survived'})
cf
```

## 4. KNN

```python
from sklearn.neighbors import KNeighborsClassifier
import math
k=math.floor(math.sqrt(X_train.shape[0]))
print(k)
knn = KNeighborsClassifier(k)
knn.fit(X_train,Y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
confusion_matrix(Y_test, y_pred)
```

```python
knn=accuracy_score(Y_test, y_pred)
knn
```

```python
print("Accuracy: ",round(accuracy_score(Y_test, y_pred)*100,2),"%")
```
